# 2. 전처리(Data Post Processing

```python
# 시간 datetime tyoe으로 변환

def change_time(f):
    l=[]
    for i in range(len(f)):
        t=str(f['time'][i])
        tmp=t[0:4]+'-'+t[4:6]+'-'+t[6:8]+' '+t[8:10]+':'+t[10:12]+':'+t[12:14]
        l.append(tmp)
        
    f['time']=l
    f['time']=pd.to_datetime(f['time'],format='%Y-%m-%d %H:%M:%S')
    
    return f
```

```python
# Support vector machine

from sklearn.svm import SVC

svm = SVC(kernel='rbf',C=10,gamma=0.1,random_state=0)
```

```python
### 교차검증

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

kfold=KFold(n_splits=5)
n_iter=0
r_accuracy=[]

for train_index, test_index in kfold.split(r_feature):
    X_train, X_test=r_feature[train_index],r_feature[test_index]
    Y_train, Y_test=r_target[train_index],r_target[test_index]
    
    svm.fit(X_train,Y_train)
    pred=svm.predict(X_test)
    n_iter+=1
    
    accuracy=np.round(accuracy_score(Y_test,pred),4)
    train_size=X_train.shape[0]
    test_size=X_test.shape[0]
    print('{0} 교차검증 정확도 :{1}, 학습 데이터 크기:{2}, 검증 데이터 크기:{3}'.format(n_iter,accuracy,train_size,test_size))
    
    r_accuracy.append(accuracy)
    
print('\n 평균검증 정확도:',np.mean(r_accuracy))
```

```python
# KNN을 사용하여

from sklearn.neighbors import KNeighborsClassifier

best_k=[0,0]
for i in range(2,11):
    knn = KNeighborsClassifier(n_neighbors=i)
    
    
    kfold=KFold(n_splits=5)
    n_iter=0
    r_accuracy=[]

    for train_index, test_index in kfold.split(r_feature):
        X_train, X_test=r_feature[train_index],r_feature[test_index]
        Y_train, Y_test=r_target[train_index],r_target[test_index]
    
        knn.fit(X_train,Y_train)
        pred=knn.predict(X_test)
        n_iter+=1
    
        accuracy=np.round(accuracy_score(Y_test,pred),4)
        train_size=X_train.shape[0]
        test_size=X_test.shape[0]
        print('{0} 교차검증 정확도 :{1}, 학습 데이터 크기:{2}, 검증 데이터 크기:{3}'.format(n_iter,accuracy,train_size,test_size))
    
        r_accuracy.append(accuracy)
        
        if best_k[0]<np.mean(r_accuracy):
            best_k[0]=np.mean(r_accuracy)
            best_k[1]=i

print('k={0} \n 평균검증 정확도:{1}'.format(best_k[1],best_k[0]))
```

```python
from sklearn.ensemble import StackingClassifier
import lightgbm


def get_stacking():
    level0=[]
    level0.append(('svm',svm))
    level0.append(('knn',KNeighborsClassifier(n_neighbors=6)))
    level0.append(('dtree',dtree))
    
    level1=lightgbm.LGBMClassifier(n_estimators=400,learning_rate=0.1,max_depth=3)
    
    model=StackingClassifier(estimators=level0, final_estimator=level1,cv=3)
    
    return model

stack=get_stacking()
```

```python
### 교차검증

from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

kfold=KFold(n_splits=5)
n_iter=0
r_accuracy=[]

for train_index, test_index in kfold.split(r_feature):
    X_train, X_test=r_feature[train_index],r_feature[test_index]
    Y_train, Y_test=r_target[train_index],r_target[test_index]
    
    stack.fit(X_train,Y_train)
    pred=dtree.predict(X_test)
    n_iter+=1
    
    accuracy=np.round(accuracy_score(Y_test,pred),4)
    train_size=X_train.shape[0]
    test_size=X_test.shape[0]
    print('{0} 교차검증 정확도 :{1}, 학습 데이터 크기:{2}, 검증 데이터 크기:{3}'.format(n_iter,accuracy,train_size,test_size))
    
    r_accuracy.append(accuracy)
    
print('\n 평균검증 정확도:',np.mean(r_accuracy))
```
